```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.ticker import MaxNLocator
```

```python
data_path = 'Data/giro-d-italia 2000-2024.csv'
df = pd.read_csv(data_path)
```

```python
scat = sns.scatterplot(data=df, x='distance', y='vertical meters',size=8, hue='time gap', legend=None)
scat.xaxis.set_major_locator(MaxNLocator(nbins=5))
scat.yaxis.set_major_locator(MaxNLocator(nbins=5))
```
